```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')

print(df.to_string())


g=sb.FacetGrid(df, col= "time")
g.map(plt.hist,"tip")
plt.show()
```

```python
g=sb.FacetGrid(df,col="sex",hue="smoker")
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
g=sb.FacetGrid(df,col="sex",hue="time")
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
sb.barplot(x="sex",y="tip",hue="time",data=df)
plt.show()
```

```python
sb.regplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
sb.pairplot(df)
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
df=sb.load_dataset('tips')
selected_vars=["total_bill","tip"]
sb.pairplot(df[selected_vars])
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
df=sb.load_dataset('tips')
selected_vars=["total_bill","tip"]
sb.pairplot(df,hue="size")
plt.show()
```

```python
#modifying color palette
import seaborn as sb
import matplotlib.pyplot as plt
df=sb.load_dataset('tips')
sb.pairplot(df,hue="size",palette="husl")
plt.show()
```

```python
#changing diagnol kind of plot
sb.pairplot(df,diag_kind="kde")
plt.show()
```

```python
sb.pairplot(df,kind="reg")
plt.show()
```

```python
#controlling the markerss
sb.pairplot(df,hue="sex",markers=["o","s"])
plt.show()
```

```python
#advanced customization with seaborm
g=sb.pairplot(df,hue="day")
g.fig.suptitle("Pairplot of tips dataset",y=1.02)
g.set(xticks=[],yticks=[])
plt.show()
```
